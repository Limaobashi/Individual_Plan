```r
library(tidyverse)
```

```r
players_data <- read_csv("data/players.csv")
players_data
```

```r
players_data_2 <- players_data |>
select(experience, subscribe, hashedEmail,played_hours, name, gender, age)
players_data_2
```

It is now a tidy dataset.

```r
players_plot <- players_data_2 |>
filter(age < 25) |>
ggplot(aes(x = age, y = played_hours)) +
geom_point(aes(colour = experience)) +
xlab("The Age of Players") +
ylab("Played Time (hours)")
players_plot
```

In that plot, I noticed that lots of data are concentrated on 0 hours, and the three players with the longest playtime all have 'Regular' experience. I think we need more data and players need to play longer. Otherwise, our prediction will face significant challenges due to a lack of sufficient data support.
